```python
import pandas as pnd
from datetime import date
import numpy as np
import operator
```

```python
data=pnd.read_csv('rideshare_kaggle.csv')
pnd.set_option('display.max_columns',None)
data.head()
```

```python
data=data.sort_values('datetime')
data=data[['id','timestamp','hour','day','month','datetime','source','destination','cab_type','product_id','name','price','distance','surge_multiplier','temperature','short_summary','long_summary','icon','visibility','temperatureHigh','temperatureLow','apparentTemperatureHigh','apparentTemperatureLow','latitude','longitude']]
data.head()
```

```python
data.info()
```

```python
data.isna().any()
```

```python
data.isnull().sum()
```

```python
data['price']=data['price'].fillna(data['distance']*1.45)
```

```python
data['source'].value_counts()
```

```python
data['destination'].value_counts()
```

```python
data["datetime"] = pnd.to_datetime(data["datetime"])
type(data["datetime"])
```

```python
data['days name']=data['datetime'].dt.day_name()
```

```python
data['time of drive'] = np.where(operator.and_(data['hour'] >= 5,data['hour']<12), 'Moring', (np.where(operator.and_(data['hour'] >= 12,data['hour']<17), 'Noon', (np.where(operator.and_(data['hour'] >= 17,data['hour']<22), 'Evening','Night')))))
```

```python
data.to_csv('uber vs lyft.csv')
data.head()
```
