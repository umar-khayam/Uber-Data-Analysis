```python
import pandas as pnd
from datetime import date
import numpy as np
import operator
```

```python
data=pnd.read_csv('rideshare_kaggle.csv')
pnd.set_option('display.max_columns',None)
data
```

```python
data=data.sort_values('datetime')
data=data[['id','timestamp','hour','day','month','datetime','source','destination','cab_type','product_id','name','price','distance','surge_multiplier','temperature','long_summary','icon','visibility','temperatureHigh','temperatureLow','apparentTemperatureHigh','apparentTemperatureLow','latitude','longitude']]
data
```

```python
data['price']=data['price'].fillna(data['price'].mean())
```

```python
data['day'].value_counts()
```

```python
data["datetime"] = pnd.to_datetime(data["datetime"])
type(data["datetime"])
```

```python
data['days name']=data['datetime'].dt.day_name()
```

```python
data['time of drive'] = np.where(operator.and_(data['hour'] >= 6,data['hour']<11), 'Moring', (np.where(operator.and_(data['hour'] >= 11,data['hour']<18), 'Noon', (np.where(operator.and_(data['hour'] >= 18,data['hour']<22), 'Evening','Late Night')))))
```

```python
data.to_csv('uber vs lyft.csv')
data
```
